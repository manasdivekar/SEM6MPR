```python
import numpy as np
import pandas as pd
```

```python
match = pd.read_csv('matches.csv')
delivery = pd.read_csv('deliveries.csv')
```

```python
match.head()
```

```python
delivery.head()
```

```python
#new DataFrame named total_score_df which contains the total runs scored in each match and inning combination.

total_score_df = delivery.groupby(['match_id','inning']).sum()['total_runs'].reset_index()
```

```python
#keeping total score of ther first inning only and removing the total score of the second inning

total_score_df = total_score_df[total_score_df['inning'] == 1]
```

```python
total_score_df
```

```python
# merging the total score dataframe (only match id and total runs attribute) with the matches dataset

match_df = match.merge(total_score_df[['match_id','total_runs']],left_on='id',right_on='match_id')
```

```python
match_df
```

```python
# Displays the team name present in the data set and now we are going to remove those teams which are currently not participating in ipl

match_df['team1'].unique()
```

```python
teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals'
]
```

```python
# Replacing the teams name with current new team name like delhi daredevils renamed into delhi capitals

match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
# only considering those data which has the following team name

match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df.shape
```

```python
# Not considering those matches were DLS system was applied so will remove that attribute from dataset
match_df = match_df[match_df['dl_applied'] == 0]
```

```python
# Considering only these four attributes and will merge this attribute with delivery dataset

match_df = match_df[['match_id','city','winner','total_runs']]
```

```python
# Merging this dataset on matchid attribute

delivery_df = match_df.merge(delivery,on='match_id')
```

```python
# considering those values for secind innings of the match

delivery_df = delivery_df[delivery_df['inning'] == 2]
```

```python
delivery_df
```

```python
# we are grouping according to that particular match and we will calculate the score after every over in that match as that is represented in total_runs_y attirbute of delivery dataset
delivery_df['current_score'] = delivery_df.groupby('match_id')['total_runs_y'].cumsum()
```

```python
# Finding out the runs left after every over as we know the target and from above cell we are able to know the current score

delivery_df['runs_left'] = delivery_df['total_runs_x'] - delivery_df['current_score']
```

```python
delivery_df['balls_left'] = 120 - (delivery_df['over']*6 + delivery_df['ball'])
```

```python
delivery_df
```

```python
# Wickets left after every over in the match is calculated here

delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x == "0" else "1")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
wickets = delivery_df.groupby('match_id')['player_dismissed'].cumsum().values
delivery_df['wickets'] = 10 - wickets
delivery_df.head()
```

```python
delivery_df.head()
```

```python
# crr = runs/overs
delivery_df['crr'] = (delivery_df['current_score']*6)/(120 - delivery_df['balls_left'])
```

```python
# calculating Required run rate
delivery_df['rrr'] = (delivery_df['runs_left']*6)/delivery_df['balls_left']
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(result,axis=1)
```

```python
# final attribute which need after performing feature selection on the dataset . Whole model is based on this attributes

final_df = delivery_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr','result']]
```

```python
# Shuffling the data
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.sample()
```

```python
final_df.dropna(inplace=True)
```

```python
final_df = final_df[final_df['balls_left'] != 0]
```

Model starts from here !!!

```python
X = final_df.iloc[:,:-1] # Every rows except the last one column
y = final_df.iloc[:,-1] # Everything
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=1)
```

```python
X_train
```

```python
# Changing the attributes having string datatype using one hot encoder

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
```

Logistic regression Pipeline

```python
pipe_lr = Pipeline(steps=[
    ('step1', trf),
    ('step2', LogisticRegression(solver='liblinear'))
])
```

```python
pipe_lr.fit(X_train,y_train)
```

```python
y_pred = pipe_lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
pipe_lr.predict_proba(X_test)[10]
```

RandomForestClassifier Pipeline

```python
pipe_rf = Pipeline(steps=[
    ('step1', trf),
    ('step2', RandomForestClassifier())
])
```

```python
pipe_rf.fit(X_train,y_train)
```

```python
y_pred = pipe_rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
pipe_rf.predict_proba(X_test)[10]
```

SUPPORT VECTOR MACHINE PIPELINE

```python
# Define the SVM model with RBF kernel
svm_classifier = SVC(kernel='rbf')
```

```python
pipe_svm = Pipeline(steps=[
    ('step1', trf),
    ('step2', svm_classifier)
])
```

```python
pipe_svm.fit(X_train,y_train)
```

```python
y_pred = pipe_svm.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
pipe_svm.predict_proba(X_test)[10]
```

```python
def match_summary(row):
    print("Batting Team-" + row['batting_team'] + " | Bowling Team-" + row['bowling_team'] + " | Target- " + str(row['total_runs_x']))
```

```python
def match_progression(x_df,match_id,pipe_lr):
    match = x_df[x_df['match_id'] == match_id]
    match = match[(match['ball'] == 6)]
    temp_df = match[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr']].dropna()
    temp_df = temp_df[temp_df['balls_left'] != 0]
    result = pipe_lr.predict_proba(temp_df)
    temp_df['lose'] = np.round(result.T[0]*100,1)
    temp_df['win'] = np.round(result.T[1]*100,1)
    temp_df['end_of_over'] = range(1,temp_df.shape[0]+1)

    target = temp_df['total_runs_x'].values[0]
    runs = list(temp_df['runs_left'].values)
    new_runs = runs[:]
    runs.insert(0,target)
    temp_df['runs_after_over'] = np.array(runs)[:-1] - np.array(new_runs)
    wickets = list(temp_df['wickets'].values)
    new_wickets = wickets[:]
    new_wickets.insert(0,10)
    wickets.append(0)
    w = np.array(wickets)
    nw = np.array(new_wickets)
    temp_df['wickets_in_over'] = (nw - w)[0:temp_df.shape[0]]

    print("Target-",target)
    temp_df = temp_df[['end_of_over','runs_after_over','wickets_in_over','lose','win']]
    return temp_df,target
```

```python
temp_df,target = match_progression(delivery_df,8,pipe_lr)
temp_df
```
